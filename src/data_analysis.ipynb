# Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the data
bottleneck_dir_name = "../data/Pedestrian_Trajectories/Bottleneck_Data/"
corridor_dir_name = "../data/Pedestrian_Trajectories/Corridor_Data/"
col_names = ["ID", "FRAME", "X", "Y", "Z"]
bottleneck = pd.read_csv(bottleneck_dir_name + "uo-180-070.txt", sep=" ", header=None, names=col_names)
corridor = pd.read_csv(corridor_dir_name + "ug-180-015.txt", sep=" ", header=None, names=col_names)
```

```python
print(bottleneck.head())
print(corridor.head())
```

```python
# save trajectories
bottleneck_trajs, corridor_trajs = {}, {}
bottleneck_ids = bottleneck['ID'].unique()
corridor_ids = corridor['ID'].unique()
for id in bottleneck_ids:
    bottleneck_trajs[id] = bottleneck[bottleneck['ID'] == id][['X', 'Y']].to_numpy()
for id in corridor_ids:
    corridor_trajs[id] = corridor[corridor['ID'] == id][['X', 'Y']].to_numpy()
```

```python
# visualize bottleneck
for id in bottleneck_ids:
    plt.scatter(bottleneck_trajs[id][:, 0], bottleneck_trajs[id][:, 1], s=1)
plt.show()
```

```python
# visualize corridor
for id in corridor_ids:
    plt.scatter(corridor_trajs[id][:, 0], corridor_trajs[id][:, 1], s=1)
# plt.scatter(corridor_trajs[16][:, 0], corridor_trajs[16][:, 1], s=1)
plt.show()
```
